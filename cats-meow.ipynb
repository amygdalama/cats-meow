```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as mpimg
from matplotlib import mlab
from scipy.io import wavfile

img = mpimg.imread("juno.JPG")
img
```

```python
plt.imshow(img)
```

```python
lum_img = img[:,:,0]
imgplot = plt.imshow(lum_img)
imgplot.set_clim(60.0,250.0)
plt.colorbar()
```

```python
plt.hist(lum_img.flatten(), 256)
plt.show()
```

```python
sgram = plt.specgram(lum_img.flatten())
```

```python
sampling_rates = [10000, 100000]
for rate in sampling_rates:
    wavfile.write('sound-%s.wav' % rate, rate, lum_img.flatten())
```
